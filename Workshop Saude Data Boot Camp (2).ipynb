```python
def to_int(lista):
    if lista == '':
        return -1
    else:
        return int(lista)
        
```

```python
def ajuste_estado(lista):
    if lista != '':
        if int(lista) == 1:
            return 'solteira'
        elif int(lista) == 2:
            return 'casada'
        elif int(lista) == 3:
            return 'viuva'
        elif int(lista) == 4:
            return 'divorciada'
        else:
            return 'ignorado'
    else:
        return 'NaN'
        
estado_civil_mae = df_2["ESTCIVMAE"].apply(ajuste_estado)
```

```python
estado_civil_mae.head()
```

```python
df_2["estado_civil_mae"] = estado_civil_mae
```

```python
df_2.head()
```

```python
cols_int = ["IDADEMAE","QTDFILVIVO"
           ,"QTDFILMORT","GRAVIDEZ","IDADEPAI"]

for i in cols_int:
    df_2[i] = df_2[i].apply(to_int)
```

```python
df_2.info()
```

```python
def ajuste_escolaridade(lista):
    if lista != '':
        if int(lista) == 1:
            return 0.0
        elif int(lista) == 2:
            return 2.0
        elif int(lista) == 3:
            return 5.5
        elif int(lista) == 4:
            return 9.5
        elif int(lista) == 5:
            return 12.0
        else:
            return -1.0
    else:
        return -1.0
    
```

```python
df_2 = df_2.drop("ESCMAE",axis=1)
```

```python
df_2["media_escolaridade_mae"] = df.ESCMAE.apply(ajuste_escolaridade)
```

```python
df_2 = df_2.drop("CODOCUPMAE",axis=1)
```

```python
def ajuste_gestacao(lista):
    if lista != '':
        if int(lista) == 1:
            return 22.0
        elif int(lista) == 2:
            return 24.5
        elif int(lista) == 3:
            return 29.5
        elif int(lista) == 4:
            return 34.0
        elif int(lista) == 5:
            return 39
        elif int(lista) == 6:
            return 42
        else:
            return -1.0
    else:
        return -1.0
    
df_2["media_gestacao"] = df.GRAVIDEZ.apply(ajuste_gestacao)
```

```python
def ajuste_parto(lista):
    if lista != '':
        if int(lista) == 1:
            return 0
        elif int(lista) == 2:
            return 1
        else:
            return -1.0
    else:
        return -1.0
df_2["parto_boolean"] = df.PARTO.apply(ajuste_parto)
```

```python
def ajuste_sexo_1(lista):
    if lista != '':
        if int(lista) == 1:
            return 0
        elif int(lista) == 2:
            return 1
        else:
            return -1.0
    else:
        return -1.0
df_2["sexo_boolean"] = df.SEXO.apply(ajuste_sexo_1)
```

```python
def ajuste_sexo_2(lista):
    if lista != '':
        if int(lista) == 1:
            return 'masculino'
        elif int(lista) == 2:
            return 'feminino'
        else:
            return 'NaN'
    else:
        return 'NaN'
df_2["sexo"] = df.SEXO.apply(ajuste_sexo_2)
```

```python
def ajuste_parto_2(lista):
    if lista != '':
        if int(lista) == 1:
            return 'vaginal'
        elif int(lista) == 2:
            return 'cesareo'
        else:
            return 'NaN'
    else:
        return 'NaN'
df_2["parto"] = df.PARTO.apply(ajuste_parto_2)
```

```python
def ajuste_consultas(lista):
    if lista != '':
        if int(lista) == 1:
            return 0
        elif int(lista) == 2:
            return 2
        elif int(lista) == 3:
            return 5
        elif int(lista) == 4:
            return 7
        else:
            return -1.0
    else:
        return -1.0
    
df_2["media_consultas"] = df.CONSULTAS.apply(ajuste_consultas)
```

```python
def ajuste_raca(lista):
    if lista != '':
        if int(lista) == 1:
            return 'branca'
        elif int(lista) == 2:
            return 'preta'
        elif int(lista) == 3:
            return 'amarela'
        elif int(lista) == 4:
            return 'parda'
        elif int(lista) == 5:
            return 'indigena'
        else:
            return 'NaN'
    else:

        return 'NaN'
df_2["raca_mae"] = df_2["RACACOR"].apply(ajuste_raca)
df_2 = df_2.drop("RACACOR",axis=1)
```

```python
def ajuste_anomalia(lista):
    if lista != '':
        if int(lista) == 1:
            return 1
        elif int(lista) == 2:
            return 0
        else:
            return -1
    else:

        return -1

df_2["anomalia_boolean"] = df.IDANOMAL.apply(ajuste_anomalia)
```
